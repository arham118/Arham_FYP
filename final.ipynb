```python
def remove_outliers_iqr(df, cols=None):
    cols = cols or df.select_dtypes(include='number').columns
    for col in cols:
        q1, q3 = df[col].quantile([0.25, 0.75])
        iqr = q3 - q1
        df = df[(df[col] >= q1 - 1.5 * iqr) & (df[col] <= q3 + 1.5 * iqr)]
    return df
merged_df = remove_outliers_iqr(merged_df)
```

```python
merged_df['month'] = pd.to_datetime(merged_df['month'])
merged_df.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# Fit KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
seller_features['cluster'] = kmeans.fit_predict(seller_features[numeric_cols])
```

```python
# Check average values per cluster
cluster_summary = seller_features.groupby('cluster').mean()
display(cluster_summary)
```

```python
#mapping clusters
cluster_labels = {
    1: 'High Growth',
    2: 'Stable Performer',
    0: 'New or Struggling Sellers'
}
```

```python
seller_features['segment'] = seller_features['cluster'].map(cluster_labels)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
pca = PCA(n_components=2)
reduced = pca.fit_transform(seller_features[numeric_cols])
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(reduced[:, 0], reduced[:, 1], c=seller_features['cluster'], cmap='viridis', alpha=0.7)
plt.title('Seller Segments Visualized with PCA')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Silhouette Score
sil_score = silhouette_score(seller_features[numeric_cols], seller_features['cluster'])
print("Silhouette Score:", sil_score)
```
